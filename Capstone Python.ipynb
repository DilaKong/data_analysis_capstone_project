```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '/Users/serhankong/Desktop/data-1726249043839.csv'
df = pd.read_csv(file_path)


plt.figure(figsize=(10, 6))
plt.pie(df['customer_count'], labels=df['city'], autopct='%1.1f%%', startangle=140)
plt.title('Şehir Bazlı Müşteri Dağılımı')
plt.show()


df_sorted = df.sort_values('customer_count', ascending=False)

plt.figure(figsize=(12, 8))
plt.bar(df_sorted['city'], df_sorted['customer_count'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Şehir Bazında Müşteri Sayıları (Azalan)')
plt.xlabel('Şehir')
plt.ylabel('Müşteri Sayısı')
plt.tight_layout()
plt.show()
```
